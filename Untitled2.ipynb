```python
X = df[['G2','G3','T5','T9', 'T30', 'R5', 'R6', 'R7','G_class', 'T_class', 'R_class','O_class']]
y = df['Y']
```

```python
from collections import Counter
print(Counter(y))
# 查看所生成的样本类别分布，1和2和3樣本數量差很多，属于类别不平衡数据
```

```python
# 使用imlbearn库中上采样方法中的SMOTE接口
from imblearn.over_sampling import SMOTE
# 定义SMOTE模型，random_state相当于随机数种子的作用
smo = SMOTE(random_state=0)
X_smo, y_smo = smo.fit_sample(X, y)
```

```python
print(Counter(y_smo))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_smo, y_smo,test_size=0.25, random_state=1234)
from sklearn.ensemble import GradientBoostingClassifier #導入梯度提升回歸樹模型
clf = GradientBoostingClassifier(random_state=1234, max_depth=7,learning_rate=1,n_estimators= 1700,subsample=1)
clf.fit(X_train,y_train)
```

```python
pred = clf.predict_proba(test)
submission = pd.read_csv('dvc_data/submit_samples.csv')
df = pd.DataFrame({ 'ID':submission.ID,'C1':pred[0:300,0] , 'C2':pred[0:300,1] , 'C3':pred[0:300,2] })
filename = 'submission_pre.csv'
df.to_csv(filename,index=False)
print(df)
print('Saved file: ' + filename)
```
